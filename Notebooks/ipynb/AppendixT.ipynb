# Appendix A:  Terminology

This Appendix summarizes the terminology used the main manuscript as relevant to:

1. Dependent variable dimensionality ($m$)
2. Measurement domain dimensionality ($n$)
3. Confidence and prediction regions

Dimensionality terminology follows Pataky et al. (2016, Table 1). Confidence/prediction region follows the general statistics literature.


##### Reference:

Pataky, T. C., Vanrenterghem, J., & Robinson, M. A. (2016). The probability of false positives in zero-dimensional analyses of one-dimensional kinematic, force and EMG trajectories. Journal of Biomechanics, 49(9), 1468â€“1476. http://doi.org/10.1016/j.jbiomech.2016.03.032

### Data dimensionality terminology

Table A1 lists terminology relevant to [dependent variables](https://en.wikipedia.org/wiki/Dependent_and_independent_variables). In the Biomechanics literature the terms "scalar" and "vector" are often used for "[univariate](https://en.wikipedia.org/wiki/Univariate)" and "[multivariate](https://en.wikipedia.org/wiki/Multivariate_statistics)". However, "vector" is ambiguous in the sense that it could represent a physical quantity (e.g. velocity vector), or it could simply represent a collection of $m$ different variables (e.g. body mass and body height). The main manuscript uses "multivariate" to encompass both possibilities, and avoids "vector" because of this ambiguity. Consequently, the main manuscript also uses "univariate" preferably over "scalar".
